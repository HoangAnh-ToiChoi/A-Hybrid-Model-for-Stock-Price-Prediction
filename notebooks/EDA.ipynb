```python
def plot_stock_data(ticker_name):
    file_path = f"../data/raw/{ticker_name}.csv"
    
    if not os.path.exists(file_path):
        print(f"Khng tm thy file {file_path}")
        return

    # 1. c file CSV
    df = pd.read_csv(file_path, index_col=0)
    
   
    df.index = pd.to_datetime(df.index, errors='coerce')
    
    df = df[df.index.notnull()]
    
    df['Close'] = pd.to_numeric(df['Close'])
 

    print(f"Clean data of {ticker_name} ")
    print(df.head())
    print("\n")

    plt.figure(figsize=(14, 6))
    plt.plot(df.index, df['Close'], label=f'{ticker_name} Close Price', color='navy')
    plt.title(f'Biu  gi ng ca - {ticker_name} (2015-2024)', fontsize=16)
    plt.xlabel('Nm', fontsize=12)
    plt.ylabel('Gi (USD)', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()
```
